```python
dataset = df["Open"]
dataset = pd.DataFrame(dataset)

data = dataset.values

data.shape
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(data)
```

```python
len(data)
```

```python
# 75% of the data fror Training and 25% of the data for Testing
train_size = int(len(data)*.75)
test_size = len(data) - train_size

print("Train size :",train_size, "Test size :",test_size)

train_data = scaled_data[ :train_size , 0:1 ]
test_data = scaled_data[ train_size-60: , 0:1 ]
```

```python
train_data.shape, test_data.shape
```

```python
# Creating a Training Set with 60 time-steps and 1 output
x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
```

```python
# Convert to numpy array
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
# Reshaping the Input
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1],1))
```

```python
x_train.shape, y_train.shape
```

```python
# Creating a LSTM model
model = Sequential()
model.add(GRU(50, return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(GRU(64, return_sequences=False))
model.add(Dense(32))
model.add(Dense(16))
model.add(Dense(1))

model.compile(optimizer="adam", loss="mse", metrics=["mean_squared_error"])
```

```python
model.summary()
```

```python
# Fitting the LSTM-Model to Training set
callbacks = [EarlyStopping(monitor="loss", patience=10, restore_best_weights=True)]
history = model.fit(x_train, y_train, epochs=100, batch_size=32, callbacks=callbacks)
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["mean_squared_error"])
plt.legend(['Mean Squared Error', "Mean Squared Error"])
plt.title("Losses")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.show()
```

```python
# Creating a Testing Set with 60 time-steps and 1 output
x_test = []
y_test = []

for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])
    y_test.append(test_data[i, 0])
```

```python
# Convert to numpy array
x_test, y_test = np.array(x_test), np.array(y_test)
```

```python
# Reshaping the Input
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
```

```python
x_test.shape, y_test.shape
```

```python
#inverse y_test scaling
predictions = model.predict(x_test)

#inverse predictions scaling
predictions = scaler.inverse_transform(predictions)
predictions.shape
```

```python
#inverse y_test scaling
y_test = scaler.inverse_transform([y_test])

RMSE = np.sqrt(np.mean( y_test - predictions )**2).round(2)
RMSE
```
